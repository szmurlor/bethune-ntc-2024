```python
# Define a function processing the dataset:
#  replace_ip2int - string representations of IP numbers are transformed to integers,
#  remove_low_occurence_protocols - remove records for protocols where we have less than 10 packets
#  remove_uninteresting_features - remove from the dataset columns which are not features

def process_dataset_df(df, low_occurance_limit=10):
    def replace_ip2int():
        def ip2int(ip):
            # Function that changes IP addresses into numbers
            packedIP = socket.inet_aton(ip)
            return struct.unpack("!L", packedIP)[0]

        df["Source.IP"] = df["Source.IP"].apply(ip2int)
        df["Destination.IP"] = df["Destination.IP"].apply(ip2int)

    def remove_low_occurence_protocols():
        nonlocal df
        # Features that will be removed from dataset because they have low occurances of records in dataset
        prot_counts = df['ProtocolName'].value_counts()
        feats_toDelete = prot_counts[ prot_counts < low_occurance_limit ].index
        feats_toDelete
        df = df[~df.ProtocolName.isin(feats_toDelete)]

    def remove_uninteresting_features():
        nonlocal df
        feat_names = df.columns.to_list()
        feat_names.remove("Label")
        to_remove = df.columns[ df.nunique() == 1 ].to_list()
        feat_names = list(filter(lambda x: x not in to_remove, feat_names))
        df = df[ feat_names ]

    replace_ip2int()
    print(f"Dimensions after ip2int: {df.shape}")
    remove_low_occurence_protocols()
    print(f"Dimensions after remove_low_occurence_protocols: {df.shape}")
    remove_uninteresting_features()
    print(f"Dimensions after remove_uninteresting_features: {df.shape}")

    return df

# here read the data from file from out_fnames[0] - the first file name
# and execute the process_dataset method on it

...
```
